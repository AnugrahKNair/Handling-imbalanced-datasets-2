```
import pandas as pd
import numpy as np
#using the cleaned data
data=pd.read_csv("framhingham.csv")
```

```
#create X and y
X=data.drop("TenYearCHD", axis=1)
y=data["TenYearCHD"]
```

```
#Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test= train_test_split(X,y, test_size=0.2, stratify=y)
```

```
print(len(X_train.columns))
```

```
n_inputs=15
```

```
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```
model=Sequential()

model.add(Dense(30, input_dim=n_inputs,activation='relu',kernel_initializer='he_uniform'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')

model.fit(X_train,y_train, epochs=1000)
```

```
from sklearn.metrics import roc_auc_score
y_pred1=model.predict(X_test)
print(roc_auc_score(y_test,y_pred1))
```

```
weights_assigned={0:1,1:18}

model=Sequential()

model.add(Dense(15, input_dim=n_inputs,activation='relu',kernel_initializer='he_uniform'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')

model.fit(X_train,y_train, class_weight=weights_assigned , epochs=100)
```

```
y_pred2=model.predict(X_test)
print(roc_auc_score(y_test,y_pred2))
```
